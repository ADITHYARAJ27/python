```python
mset={1,2,3}
print(mset)
```

```python

mset.add(4)
print(mset)
```

```python
mset.update([4,5,6,7])
print(mset)
```

```python
mset.remove(2)
print(mset)
```

```python
#an element which is not in this will show an error

mset.remove(9)
print(mset)
```

```python
#where in discard it will not show any error message

mset.discard(9)
print(mset)
```

```python
mset.discard(6)
print(mset)
```

```python
#pops a randon number from the list

poppeditem=mset.pop()
print(poppeditem)
print(mset)
```

```python
#clears all the elemetns in set

mset.clear()
print(mset)
```

```python
#union

set1={1,2,3,4}
set2={4,5,6,7}

unionset= set1 | set2
print(unionset)
```

```python
#intersection

inte= set1 & set2
print(inte)
```

```python
diff = set1 - set2
print(diff)
```

```python
#symmetric difference

sd= set1 ^ set2
print(sd)
```

```python
#subset and superset

#set1 = 1-100
#set2 = 20-60

#set1 is a superset of set2
#set2 is a subset of set1
```

```python
set1 = {1,2,3,4,5}
set2 = {2,3}

print(set1.issubset(set2))
print(set1.issuperset(set2))
print(set2.issubset(set1))
```

```python
#disjoint set -  sets which doesnt contain any common elements.

set1={1,2,3,4}
set2={5,6,7}

print(set1.isdisjoint(set2))
```
